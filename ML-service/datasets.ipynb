Merging

```python
df = train_df.copy()
```

```python
df = df.merge(items_df, on='item_nbr', how='left')
df.head()
```

```python
df = df.merge(stores_df, on='store_nbr', how='left')
df.head()
```

```python
df = df.merge(oil_df, on="date", how="left")
df["dcoilwtico"] = df["dcoilwtico"].ffill()
df.head()
```

```python
df = df.merge(transactions_df, on=["date", "store_nbr"], how="left")
df.head()
```

```python
holidays_events_df
```

```python
holidays_events_df = holidays_events_df[holidays_events_df['transferred'] == False]
holidays_events_df
```

```python
holidays_events_df = holidays_events_df.groupby("date").agg({
    "type": "first",
    "locale": "first"
}).reset_index()
holidays_events_df["is_holiday"] = 1
holidays_events_df
```

```python
df = df.merge(holidays_events_df[["date", "is_holiday"]], on='date', how='left')
df["is_holiday"] = df["is_holiday"].fillna(0)
df.head()
```

```python
df.info(show_counts=True)
```

```python
cols_to_keep = [
    "date",
    "store_nbr",
    "item_nbr",
    "unit_sales",
    "cluster",
    "state",
    "is_holiday"
]

df = df[cols_to_keep]
```

```python
df["date"] = pd.to_datetime(df["date"])
print(df["date"].min())
print(df["date"].max())
```

```python
df = df[df["date"] >= '2017-01-01']
df.info(show_counts=True)
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df = df.drop(columns=["index"])
df.head()
```

```python
df = df[df["date"] <= '2017-6-30']
df.info(show_counts=True)
```

```python
df["item_nbr"].nunique()
```

```python
df["store_nbr"].nunique()
```

```python
combo_counts = (
    df.groupby(["store_nbr", "item_nbr"], as_index=False)
    .size()
    .rename(columns={"size": "record_count"})
    .sort_values("record_count", ascending=False)
)

print(combo_counts.describe())
print(f"\nTotal unique combinations: {len(combo_counts):,}")
print(f"Expected max records per combo: {df['date'].nunique()}")
```

```python
small_df = df[df["store_nbr"].isin([1, 2, 3])]
print(f"Rows: {len(small_df):,}")
print(f"Stores: {small_df['store_nbr'].unique()}")
print(f"Memory: {small_df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
```

```python
daily_sales = df.groupby('date').agg(
    unit_sales=('unit_sales', 'sum'),
    item_count=('item_nbr', 'nunique')
).reset_index()
daily_sales
```
